```python
data.drop(columns=['CARR','CTVA','DOW','FOXA','FOX','LUMN','OGN','OTIS'], inplace=True)
```

```python
missing_values = data.isnull().sum()
print(missing_values[missing_values > 0])
```

```python
for column in data.columns:
    if data[column].isnull().any():
        mean_value = data[column].mean()
        data[column].fillna(mean_value, inplace=True)

print(data.isnull().sum())
```

```python
data.duplicated().sum()
```

```python
data['Date'] = pd.to_datetime(data['Date'])
```

```python
data['Period'] = ['Before' if date < pd.Timestamp('2020-03-01') else 'After' for date in data['Date']]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
features = data.select_dtypes(include=['float64', 'int64'])
```

```python
scaler = StandardScaler()
standardized_data = scaler.fit_transform(features)
```

```python
standardized_df = pd.DataFrame(standardized_data, columns=features.columns)
```

```python
non_numeric_columns = data.select_dtypes(exclude=['float64', 'int64'])
final_df = pd.concat([non_numeric_columns.reset_index(drop=True), standardized_df.reset_index(drop=True)], axis=1)

print(final_df.head())
```

```python
# Univariate Analysis (Distributions)
def univariate_analysis(final_df, cols):
  for col in cols:
      if col in final_df.columns:  # Check if column exists before plotting
          plt.figure(figsize=(10, 6))
          final_df[col].hist()
          plt.xlabel(col)
          plt.ylabel("Frequency")
          plt.title(f"Distribution of {col}")
          plt.show()

# Select columns for analysis (replace with your desired columns)
cols_to_analyze = ['MMM', 'Low', 'ABT', 'ABBV','ABMD']
univariate_analysis(final_df, cols_to_analyze)
```

```python
#Distribution Plots for the first 10 features

first_10_cols = data_dropped.columns[:10]
for col in first_10_cols:
    plt.figure(figsize=(8, 4))
    sns.histplot(final_df[col].dropna(), bins=30, kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
import pandas as pd

# Assuming 'df' is your dataframe
df_long = pd.melt(final_df, id_vars=["Date", "Period"], var_name="Company", value_name="close")

# Show the reshaped data
df_long.tail(100)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Group by Date and Period, then calculate the average stock price for each group
df_grouped = df_long.groupby(['Date', 'Period'])['close'].mean().reset_index()

# Plot the results
plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='close', hue='Period', data=df_grouped)
plt.title('Average Stock Price (Before vs After)')
plt.xlabel('Date')
plt.ylabel('Average Stock Price')
plt.legend(title='Period')
plt.xticks(rotation=45)
plt.show()
```
